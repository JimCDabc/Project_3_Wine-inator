```python
import pandas as pd 
import os
import matplotlib.pyplot as plt  
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
```

```python
red_wine_csv = os.path.join("..", "..", "datasets", "winequality-red.csv")
red_wine_df = pd.read_csv(red_wine_csv, delimiter=";")
red_wine_df.head()
```

```python
red_wine_df.describe()
```

```python
red_wine_df.plot(x='fixed acidity', y='quality', style='o')  
plt.title('fixed acidity')  
plt.xlabel('fixed acidity')  
plt.ylabel('quality')  
plt.xlim(3.5,16.5)
plt.show()

red_wine_df.plot(x='residual sugar', y='quality', style='o')  
plt.title('residual sugar')  
plt.xlabel('residual sugar')  
plt.ylabel('quality')  
plt.xlim(0,16.5)
plt.show()
```
