```python
import pandas as pd
from pygeocoder import Geocoder
import numpy as np
from config import api_key as google_api
```

```python
wine_file = "winerygeodata.csv"
wine_df = pd.read_csv(wine_file)
wine_df.head()
```

```python
wine_df["full_address"]= wine_df["address"] + ", " + wine_df["city/zip"] + ", USA"
                                                      
                                                
```

```python
wine_df.head()
```

```python
address = wine_df["full_address"]
results = Geocoder(google_api).geocode(address)
coords = results[0].coordinates
lat = coords[0]
lon = coords[1]
```

```python
def geo_AddressToLatLon(address) : 
    ## geocode call to get lat, lon coordinates from address
    results = Geocoder(google_api).geocode(address)
    
    return (results.coordinates[0], results.coordinates[1])
                
```

```python
results = Geocoder(google_api).geocode("502 E 560th Rd, Walnut Grove, MO 66770, USA")
results.coordinates
```

```python
wine_df[["lat", "lon"]] = wine_df.apply(lambda x: pd.Series(geo_AddressToLatLon(x.full_address)), axis=1)

wine_df.head()
```
